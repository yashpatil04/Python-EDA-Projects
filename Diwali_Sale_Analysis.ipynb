# Basic EDA

```python
df.columns
```

# Gender

```python

ax = sns.countplot(x = "Gender", data = df)  # You can also write ax = sns.countplot(x = df["Gender"])

# To add count a Gender we use for loop
for bar in ax.containers:
    ax.bar_label(bar)
```

# Insight: Females placed the most orders (7,828) compared to males (3,403).S

```python
sales_gain = df.groupby(["Gender"],as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)
sns.barplot(x = "Gender", y = "Amount", data = sales_gain )
plt.show()
```

# Insight: Females spent more money than males, with the total purchase amount significantly higher for females. 

# Age

```python
df.columns
```

```python
ax = sns.countplot(x = "Age Group", data = df, hue = "Gender")

# To count the people by gender
for bar in ax.containers:
    ax.bar_label(bar)
```

# Insight: The age group 26–35 placed the most orders, with females leading in this category.

```python
# Total amount Vs Age group 
sales_age = df.groupby(["Age Group"], as_index = False )["Amount"].sum().sort_values(by = "Amount", ascending = False)
sns.barplot(x = "Age Group", y = "Amount" , data = sales_age)
plt.show()
```

# Insight: The age group 26–35 females spent the most money, contributing the highest purchase amount among all groups.

# State 

```python
df.columns
```

```python
# Total number of oders from top 10 region
state_orders = df.groupby(["State"], as_index = False)["Orders"].sum().sort_values(by = "Orders", ascending = False).head(10)

sns.set(rc ={ "figure.figsize" : (20,10)})                       # To adjust the label name (x axis)
sns.barplot(x = "State", y = "Orders", data = state_orders, palette = "tab10")
plt.show()
```

# Insight: The top 10 states with the most orders are shown, with Uttar Pradesh leading, followed by Maharashtra and Karnataka.

```python
# Amounts VS Orders 
state_amount = df.groupby("State", as_index = False)["Amount"].sum().sort_values( by = "Amount", ascending = False).head(10)

sns.set(rc ={ "figure.figsize" : (20,10)})                       # To adjust the label name (x axis)
sns.barplot(x = "State", y = "Amount", data = state_amount)
plt.show()
```

# Insight: The top 10 states with the highest purchasing power are shown, with Uttar Pradesh spending the most, followed by Maharashtra and Karnataka.

# Maritial Status

```python
ax = sns.countplot(data = df, x = "Marital_Status")

sns.set(rc ={ "figure.figsize" : (5,5)})                       # To adjust the label name (x axis)
for bat in ax.containers:
    ax.bar_label(bar)
plt.show()
```

# Insight: Married people placed significantly more orders (1,270 orders) compared to unmarried people (134 orders).

```python
sales_state = df.groupby(["Marital_Status", "Gender"],as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.set(rc ={ "figure.figsize" : (5,5)})                       # To adjust the label name (x axis)
sns.barplot(data = sales_state, x = "Marital_Status", y = "Amount" , hue = "Gender")
plt.show()
```

#  Insight: In the Marital_Status category, females spent the most money compared to males.

# Occupation

```python
# making the count plot to count which Occupation place most order
sns.set(rc ={ "figure.figsize" : (20,5)})                       # To adjust the label name (x axis)
ax = sns.countplot(data = df, x = "Occupation", palette= "tab10")

for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

# Insight: The top 3 occupations placing the most orders are Healthcare, Government, and Automobile.

```python
gp = df.groupby("Occupation", as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)


sns.set(rc ={ "figure.figsize" : (20,18)})                       # To adjust the label name (x axis)
sns.barplot(data = gp , x = "Occupation", y = "Amount", palette="tab10" )
plt.show()
```

# Insight: The top 3 occupations that spent the most amount are IT Sector, Healthcare, and Aviation.

# Product_Category

```python
df.columns
```

```python
sns.set(rc ={ "figure.figsize" : (25,25)})                       # To adjust the label name (x axis)
ax = sns.countplot(data = df , x = "Product_Category")

for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

# Insight: In the product category, Clothing & Apparel received the most orders, followed by Food and Electronics & Gadgets.

```python
gp = df.groupby("Product_Category", as_index = False)["Amount"].sum().sort_values(by = "Amount", ascending = False)

sns.set(rc ={ "figure.figsize" : (30,30)})                       # To adjust the label name (x axis)
sns.barplot(data = gp, x = "Product_Category", y = "Amount")
plt.show()
```

# Insight: Most money was spent on Food, followed by Clothing & Apparel and Electronics & Gadgets.
